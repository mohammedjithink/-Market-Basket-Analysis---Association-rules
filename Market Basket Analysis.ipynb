```python
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_csv('D:/python learning/online_retail_II.csv')
```

```python
df
```

```python
#data preparation
```

```python
df['Description']=df['Description'].str.strip()
df.dropna(axis=0, subset=['Invoice'],inplace=True)
df['Invoice']=df['Invoice'].astype('str')
df
```

```python
df['Country'].value_counts()
```

```python
#seperate transactions for germany

my_basket=(df[df['Country']=='Germany'].groupby(['Invoice','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Invoice'))
```

```python
my_basket
```

```python
#define a small function to convert values to 1 and everything else to 0

def my_encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
my_basket_sets=my_basket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE',inplace=True,axis=1)
```

```python
my_basket_sets
```

```python
## Training the Model
```

```python
#Generate the frequent item sets 
my_frequent_itemsets=apriori(my_basket_sets,min_support=0.07,use_colnames=True)
```

```python
#generating rules

my_rules=association_rules(my_frequent_itemsets, metric="lift",min_threshold=1)
```
